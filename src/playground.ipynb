```python
from helpers import *
import numpy as np
import matplotlib.pyplot as plt

from src.DiscriminatorModel import DiscriminatorModel
```

```python
sprite_type = 'people'

data = load_pixel_art_training_data('rgb', sprite_type)
uniques = np.unique(data, axis=0)
```

```python
print(uniques.shape)
```

```python
width = 25
height = 30

fig, ax = plt.subplots(nrows=height, ncols=width, figsize=(width, height))
plt.suptitle(sprite_type, fontsize=48)
for i in range(width*height):
    plt.subplot(height, width, i + 1)
    if i < uniques.shape[0]:
        plt.imshow(np.clip(uniques[i], 0, 1), cmap='gray', interpolation='nearest')
    plt.axis('off')

plt.savefig(f'{sprite_type}-example.png')
plt.show()
```
